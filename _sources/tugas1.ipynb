# 1. Statistik Deskrit

## **1. Membaca Data Dari URL**

```python
#Import Dataset
import pandas as pd
url = "https://vincentarelbundock.github.io/Rdatasets/csv/dragracer/rpdr_contestants.csv"
dataset = pd.read_csv(url)
```

```python
dataset.head(10)
```

```python
dataset.info()
```

```python
data = dataset["age"]
data
```

## **2. Membuat statistik dari data setiap atribut yang bertipe**

### **2.1 Rata-rata (Mean)**

```python
dataset["age"].mean()
```

### **2.2 Nilai sering muncul (Modulus)**

```python
dataset["age"].mode()
```

### **2.3 Nilai Tengah (Median)**

```python
dataset["age"].median()
```

## **3. Mendeskripsikan kemiringan / skwened dari data setiap variabel**

```python
data.skew()
```
