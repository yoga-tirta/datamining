## **2. Membuat statistik dari data setiap atribut yang bertipe Numerik**

Untuk membuat statistik dari dataset kita, kita membutuhkan data yang bertipe *Numerik*, maka kita tampilkan data yang bertipe Numerik seperti berikut :

### **2.1 Rata-rata (Mean)**

**Mean** merupakan angka representasi dari suatu kelompok data yang mewakili data secara keseluruhan. Menentukan mean dapat dilakukan dengan cara menjumlahkan seluruh nilai data, kemudian membaginya dengan banyaknya data.

>Rumus dari Mean yaitu

$$
      mean = \frac{1}{n} \sum_{i=i}^{n} x_{i}
$$

```python
# Menentukan Mean
dataset["age"].mean()
```

### **2.2 Nilai Tengah (Median)**

**Median** memiliki pengertian sebagai nilai data tengah. Menentukan Median dapat dilakukan dengan cara semua data  kita urutkan dari yang terkecil sampai terbesar kemudian kita ambil nilai tengahnya

>Rumus dari Median yaitu

$$
\mathcal {median} = tb + (\frac {\frac {n} 2 - F } {Fm} ) p
$$



- Tb = tepi bawah dari kelas n/2
- F = Frekuensi kumulatif sebelum kelas median
- Fm = frekuensi kelas median
- p = interval

```python
# Menentukan Median
dataset["age"].median()
```

### **2.3 Nilai sering muncul (Modulus)**

**Modulus** adalah nilai yang paling sering muncul dalam suatu data. Modus juga merupakan nilai mayoritas atau nilai dengan frekuensi paling tinggi

```python
# Menentukan Modulus
dataset["age"].mode()
```

### **2.4 Kuartil**

**Kuartil** adalah nilai yang membagi data yang berurutan menjadi empat bagian yang sama banyak. Karena data terbagi menjadi empat bagian yang sama, artinya terdapat tiga nilai kuartil, yaitu kuartil bawah (Q1), kuartil tengah (Q2), dan kuartil atas (Q3).

```python
# Menentukan Kuartil
dataset["age"].quantile()
```

## **3. Mendeskripsikan kemiringan / skwened dari data setiap variabel**

**Skewness** adalah tingkat ketidaksimetrisan atau kejauhan simetri dari sebuah distribusi. Skewness diartikan sebagai kemiringan distribusi data.

>Rumus dari Skewness yaitu

$$
skewness =\frac{\sum_{i=1}^{n}\left(x_{i}-\bar{x}\right)^{3}}{(n-1) s^{3}}
$$


- $\bar{x}$ adalah sampel mean
- s adalah standar deviasi
- $x_{i}$ adalah sampel

```python
data.skew()
```
