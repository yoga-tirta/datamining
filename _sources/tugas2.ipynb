# **Bining data**

Binning adalah cara untuk mengelompokkan suatu data ke dalam jumlah yang lebih kecil pada *“bins”*. Contohnya : jika Anda memiliki data mengenai sekelompok orang yang memiliki berbagai macam usia, Anda mungkin ingin mengatur usia mereka kedalam interval usia yang lebih sedikit, kamu dapat menggunakan metode binning pada TIBCO Spotfire untuk mengecilkan spesifikasi usia pada anggota kelompok tersebut.

Terdapat juga opsi lain dalam mengelompokkan data-data yang sudah dikategorikan ke dalam *“bins”*. Metode ini sangat membantu ketika Anda memiliki banyak kategori data dalam kolom daripada yang seharusnya dibutuhkan, sehingga Anda dpat menyederhanakannya melalui ini.

## **1. Lakukan proses diskritisasi dengan equal width dan equal frequency**



### **1.1 Equal Width**

> Equal-width intervals adalah discretization yang membagi data numerik menjadi beberapa kelompok dengan lebar kelompok yang kurang lebih sama besar.

Tahap pertama membuat kategori








```python
#Kategori
kategori = ['sedikit','sedang','banyak']
```

Kemudian menggunakan fungsi linspace() untuk membuat array yang berisi sejumlah angka dengan jarak yang sama.

```python
#Menarik batas dengan 4 kunci
bins = np.linspace(min(dataset['yield']), max(dataset['yield']), 4)
bins
```

Selanjutnya menggunakan fungsi cut() untuk melakukan binning

```python
dataset['ukuran_binned_equal'] = pd.cut(dataset['yield'], bins, labels=kategori, include_lowest=True)
```
