## **1. Lakukan proses diskritisasi dengan equal width dan equal frequency**

### **1.1 Equal Width**

Equal-width intervals adalah discretization yang membagi data numerik menjadi beberapa kelompok dengan lebar kelompok yang kurang lebih sama besar.

>Tahap pertama membuat kategori








```python
#Kategori
sepal_length_eqW  = dataset[["sepal.length"]]
sepal_width_eqW   = dataset[["sepal.width"]]
petal_length_eqW  = dataset[["petal.length"]]
petal_width_eqW   = dataset[["petal.width"]]
```

>Kemudian membuat fungsi cut yang digunakan untuk mencaari nilai interval dari metode Equal-Width-Intervals yang disediakan Pandas

```python
def cut(col, k):
  intervals = pd.cut(dataset[col], k).value_counts().index.to_list()
  return [[interval.left, interval.right] for interval in intervals]
```

```python
def toCategory(list_interval, col):
  # get length interval
  length = len(list_interval)

  # sorting interval
  sort_interval = np.sort(list_interval, axis=0)

  # get category from interval
  categories = np.array([chr(65+i) for i in range(length)])[:, None]

  # Combine into interval data
  intervals = np.hstack((sort_interval, categories))

  # operate all data
  newCol = []
  for i, row in dataset.iterrows():
    d = row[col]
    for interval in intervals:
      if d >= interval[0].astype(float) and d <= interval[1].astype(float):
        newCol.append(interval[2])
        break

  # return new column category
  return np.array(newCol, dtype=str)
```

> Kemudian mencari interval dengan membaginya menjadi 3 bagian

```python
interval_sepal_length = cut("sepal.length", 3)
interval_petal_length = cut("petal.length", 3)
interval_sepal_width = cut("sepal.width", 3)
interval_petal_width = cut("petal.width", 3)

print("Interval sepal.length =", interval_sepal_length)
print("Interval petal.length =", interval_petal_length)
print("Interval sepal.width  =", interval_sepal_width)
print("Interval petal.width  =", interval_petal_width)
```

> Lalu kita tampilkan hasil dari pembagian category

```python
sepal_length_eqW["category"] = toCategory(interval_sepal_length, "sepal.length")
petal_length_eqW["category"] = toCategory(interval_petal_length, "petal.length")
sepal_width_eqW["category"] = toCategory(interval_sepal_width, "sepal.width")
petal_width_eqW["category"] = toCategory(interval_petal_width, "petal.width")

display(sepal_length_eqW)
display(petal_length_eqW)
display(sepal_width_eqW)
display(petal_width_eqW)
```

### **1.2 Equal Frequency Intervals**

Equal Frequency Intervals adalah discretization yang membagi data numerik menjadi beberapa kelompok dengan jumlah anggota yang kurang lebih sama besar.

>Tahap pertama membuat kategori

```python
sepal_length_fr = dataset[["sepal.length"]]
petal_length_fr = dataset[["petal.length"]]
sepal_width_fr = dataset[["sepal.width"]]
petal_width_fr = dataset[["petal.width"]]
```

> Kemudian membuat fungsi cut yang digunakan untuk mencaari nilai interval dari metode Equal-Frequency-Intervals yang disediakan Pandas

```python
def qcut(col, k):
  intervals = pd.qcut(dataset[col], k).value_counts().index.to_list()
  return [[interval.left, interval.right] for interval in intervals]
```

> Kemudian mencari interval dengan membaginya menjadi 3 bagian

```python
interval_sepal_length = qcut("sepal.length", 3)
interval_petal_length = qcut("petal.length", 3)
interval_sepal_width = qcut("sepal.width", 3)
interval_petal_width = qcut("petal.width", 3)

print("Interval sepal.length =", interval_sepal_length)
print("Interval petal.length =", interval_petal_length)
print("Interval sepal.width  =", interval_sepal_width)
print("Interval petal.width  =", interval_petal_width)
```

> Lalu kita tampilkan hasil dari pembagian category

```python
sepal_length_fr["category"] = toCategory(interval_sepal_length, "sepal.length")
petal_length_fr["category"] = toCategory(interval_petal_length, "petal.length")
sepal_width_fr["category"] = toCategory(interval_sepal_width, "sepal.width")
petal_width_fr["category"] = toCategory(interval_petal_width, "petal.width")

display(sepal_length_fr)
display(petal_length_fr)
display(sepal_width_fr)
display(petal_width_fr)
```

### **2. Entrophy**

**Entrophy** adalah nilai informasi yang menyatakan ukuran ketidakpastian(impurity) dari attribut dari suatu kumpulan obyek data dalam satuan bit. Tujuan dari algoritma ini adalah mendapatkan partisi yang mengandung baris data dari kelas yang sama sebanyak mungkin.

>Rumus dari Entrophy yaitu
$$
\begin{align*}
\displaystyle Entropy(S) &= \sum_{i=0}^{k} -pi \ log_{2} \ pi \\
\end{align*}
$$



> Menambah sample data

```python
sample = dataset[["sepal.length"]]
sample.describe()
```

> Membuat kategori random untuk sample data

```python
np.random.seed(0)
sample["kategori"] = np.where(np.random.choice(2,sample.shape[0]) < 1,"A","B")
sample
```

> Membuat fungsi *GetOverollCategory* untuk menghitung data keseluruhan yang nantinya digunakan untuk menghitung entrophy

```python
def GetOverollCategory(col):
  group = sample.loc[:, :].groupby("kategori").count()
  a = group.loc["A", col]
  b = group.loc["B", col]
  return (a,b,a+b)
```

> Membuat fungsi splitter yang digunakan untuk membuat split antara value yang telah ditentukan lalu mengembalikan data yang telah dipisahkan

```python
def splitter(value, col):
  #get data less and greater from value
  less = sample[sample[col] <= value]
  greater = sample[sample[col] > value]

  #calculate into category for each data
  less_group = less.loc[:, :].groupby("kategori").count()
  greater_group = greater.loc[:, :].groupby("kategori").count()

  #get value based on category
  less_category_A = less_group.loc["A",col]
  less_category_B = less_group.loc["B",col]
  greater_category_A = greater_group.loc["A",col]
  greater_category_B = greater_group.loc["B",col]

  return (
      [less_category_A, less_category_B, less_category_A + less_category_B],
      [greater_category_A, greater_category_B, greater_category_A + greater_category_B]
  )
```

> Membuat fungsi untuk melakukan perhitungan entrophy

```python
def entropy (d):
  r1 = (d[0] / d[2]) * np.log2(d[0] / d[2])
  r2 = (d[1] / d[2]) * np.log2(d[1] / d[2])
  return np.sum([r1, r2]) * -1
```

> Membuat fungsi Info dengan rumus
$$
Info_A(D) = \frac{|D_1|}{|D|} Entropy (D_1)\frac{|D_2|}{|D|}+ Entropy (D_2)
$$

```python
def info(d):
  r1 = (d[0][2] / sample.shape[0]) * entropy(d[0])
  r2 = (d[1][2] / sample.shape[0]) * entropy(d[1])
  return r1 + r2
```

> Membuat fungsi Gain untuk menghitung selisih antara entropy awal dengan yang baru

> Rumus mencari Gain :
$$
Gain(E_{new}) = (E_{initial}) \ - (E_{new})
$$

```python
def gain(Einitial, Enew):
  return Einitial - Enew
```

> Membuat Dinitial

```python
D = GetOverollCategory("sepal.length")
entropy_d = entropy(D)
print(D)
print(entropy_d)
```

Melakukan beberapa fungsi split untuk mendapatkan informasi yang terbaik

> Tes Pertama : Split 4.5

```python
split1 = splitter(4.5, "sepal.length")
info_split1 = info(split1)
gain(entropy_d, info_split1)
```

> Tes Kedua : Split 5.5

```python
split2 = splitter(5.5, "sepal.length")
info_split2 = info(split2)
gain(entropy_d, info_split2)
```

> Tes Ketiga : Split 3.5

```python
split3 = splitter(6.5, "sepal.length")
info_split3 = info(split3)
gain(entropy_d, info_split3)
```

Diperoleh hasil split terbaik pada **Split 3** yang memberikan keuntungan informasi sebesar *0.00012921701793400242*, karena hasil tes Split 3 memiliki nilai split paling rendah.
