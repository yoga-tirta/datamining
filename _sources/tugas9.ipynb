Membuat Regresi Linier dengan menggunakan 1 fitur

```python
from sklearn.datasets import make_regression
from matplotlib import pyplot
import numpy as np
```

```python
# Membuat data dengan 1 fitur

X_test, y_test = make_regression(n_samples=100, n_features=1, noise=10)
pyplot.scatter(X_test, y_test)
pyplot.show()
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_test, y_test)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(X_test, y_test)
```

```python
pyplot.scatter(X_test, y_test)
y_plot = reg.coef_*X_test+reg.intercept_

pyplot.plot(X_test, y_plot, color ='r')
pyplot.show()
```

```python
reg.predict([[3]])
```
